```python
#!pip install crochet
import scrapy
from scrapy.crawler import CrawlerProcess
from scrapy.utils.project import get_project_settings
import pandas as pd
# from crochet import setup, wait_for
# setup()
```

```python
class BooksToScrapeSpider(scrapy.Spider):
    name = 'bookspider'

    #Website spider sends requests to
    start_urls = ['https://books.toscrape.com/']
    
    # This will help to save the data generated
    custom_settings = {
        'FEEDS': {
            'books.csv': {
                'format': 'csv',
                'overwrite': True
            }
        }
    }

    def parse(self, response):
        print('[ OUR RESPONSE ]')

        # Use the CSS selector to get the appropriate tag from the website it is in h3 and a
        #print(response.css("h3 a::text").get())
        #print(response.css("article"))
        
        ebooks = response.css("article")
        
        for ebook in ebooks:
            title = ebook.css("a::text").get()
            price = ebook.css("p.price_color::text").get()
            price = float(price[1:])
            link = ebook.css("a::attr(href)").get()
            yield {
                'Link':link,
                'Title': title,
                "Price": price
                
            } 
```

```python
#@wait_for(10)
def run_spider():
    process = CrawlerProcess(settings=get_project_settings())
    process.crawl(BooksToScrapeSpider)
    process.start()

run_spider()
```
